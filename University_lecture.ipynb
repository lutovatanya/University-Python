# Декораторы
2:39 (18.10)


```python
def my_decorator(func):
    def new_funk(*args, **kwargs):
        print('Start ' + func.__name__)
        print(' args ', args) #позиционные аргументы функции
        print(' kwargs ', kwargs)
        res = func(*args, **kwargs)
        print('Finish ' + func.__name__ + ' Result ' + str(res))
        return res
    return new_funk
```

```python
@my_decorator
def funk1(a, b, c = 5 ):
    return a * b + c
```

```python
print(funk1(1,2, c = 0))
```

```python
a = my_decorator(funk1)
a(1, 2)
```

# Property

```python
class Table:
    def __init__(self):
        self._color = '#000000'
    
    def get_color(self):
        return self._color
    
    def set_color(self, color):
        if len(color) != 7:
            print('color != 7')
            raise ValueError
        self._color = color
```

```python
table = Table()
print(table.get_color())
table.set_color('#FFFFFF')
print(table.get_color())
```

```python
table.set_color('#FFFFFFF')
print(table.get_color())
```

```python
class Table:
    def __init__(self):
        self._color = '#000000'
    
    @property
    def color(self):
        return self._color
    
    @color.setter
    def color(self, color):
        if len(color) != 7:
            print('color != 7')
            raise ValueError
        self._color = color 
    
```

```python
table = Table()
print(table.color)
table.color = '#FFFFFF'
print(table.color)
```

# Iterators
Знаем текущее состояние, знаем следующее состояние


```python
class MyIterator:
    def __init__(self, n):
        self.n = n
        pass
    def __iter__(self): #готовит объект к началу операции
        self.element_id = self.n
        return self
    def __next__(self):
        if self.element_id == 0:
            raise StopIteration
        else:
            cur = self.element_id
            self.element_id -= 1
            return cur, cur * cur, cur ** 3
```

```python
it = MyIterator(10)
it = iter(it)
for i in it:
     print(i)
```

```python
it = MyIterator(10)
it = iter(it)
i, j, k = next(it)
print(i, j, k)
i, j, k = next(it)
print(i, j, k)
```

```python
it = MyIterator(10)
it = iter(it)
for i in range(9):
    print(next(it))
```

```python
it = MyIterator(10)
for i in it:
    print(next(it))
```

```python
import time
class InvocationLoggerMetaClass(type):
     def __new__(cls, classname, parents, attributes):
        new_attributes = {}
        for key, attr in attributes.items():
            if callable(attr):
                def make_logger(key, attr):
                    def logger(*args, **kwargs):
                        print('Start method', key)
                        be = time.time()
                        res = attr(*args, **kwargs)
                        en = time.time()
                        print('Finish', key, 'Time', en - be)
                    return logger
                attr = make_logger(key, attr)
            new_attributes[key] = attr
            
            
        return super().__new__(cls, classname, parents, new_attributes)
```

```python
class Account(object, metaclass = InvocationLoggerMetaClass):
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def compute_sum(self):    
        return self.x + self.y + self.z
    def compute_mul(self):
        return self.x * self.y * self.z
    def compute_max(self):
        return max(self.x, self.y, self.z)
```

```python
acc = Account(1, 2, 3)
acc.compute_sum()
```
